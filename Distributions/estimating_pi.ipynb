```python
import numpy as np
import matplotlib.pyplot as plt 

n = 20000 # 20,000 tosses 
x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)

# plot the running estimate of Ï€, and the points in the circle
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
Ï€ = []
count = 0
for i in range(n):
    if x[i]**2 + y[i]**2 <= 1: # x^2 + y^2 <= 1 is the equation of a circle
        count += 1 # 
        ax[0].plot(x[i], y[i], 'ro')
    else:
        ax[0].plot(x[i], y[i], 'bo')
    Ï€.append(4 * count / (i+1)) # running estimate of Ï€
    ax[1].plot(range(i+1), Ï€, 'k-')

ax[0].set_title('Points in the circle')
ax[1].set_title('Running estimate of Ï€')

# estimate Ï€
Ï€ = 4 * count / n
print('Ï€ = {}'.format(Ï€))
print('The number of points in the circle is {}'.format(count))
count = 0 
```
